#Culture

## English-Hausa Pair on Culture

```python
! rm ha.enc en.enc

! python3 ${LASER}/source/embed.py \
    --encoder "/content/LASER/models/bilstm.93langs.2018-12-26.pt" \
    --bpe-codes "/content/LASER/models/93langs.fcodes" \
    --input 'English_Culture.txt' \
    --output 'en.enc' \
    --verbose 

! python3 ${LASER}/source/embed.py \
    --encoder "/content/LASER/models/bilstm.93langs.2018-12-26.pt" \
    --bpe-codes "/content/LASER/models/93langs.fcodes" \
    --input 'Hausa_Culture.txt' \
    --output 'ha.enc' \
    --verbose 
```

```python
import numpy as np
dim = 2048

en_emb = np.fromfile("en.enc", dtype=np.float32, count=-1)                                                                          
en_emb.resize(en_emb.shape[0] // dim, dim)                                                                                                 
print(en_emb.shape)

ha_emb = np.fromfile("ha.enc", dtype=np.float32, count=-1)                                                                          
ha_emb.resize(ha_emb.shape[0] // dim, dim)                                                                                                 
print(ha_emb.shape)
```

```python
! python3 ${LASER}/source/mine_bitexts.py \
      'English_Culture.txt' 'Hausa_Culture.txt' \
       --src-lang 'en' --trg-lang 'ha' \
       --src-embeddings 'en.enc' --trg-embeddings 'ha.enc' \
       --unify --mode mine --retrieval max --margin ratio -k 4  \
       --output "Ha-En_Culture_candidates.tsv" \
       --verbose
```

```python
import pandas as pd
pair=['Score', 'English', 'Hausa']
df=pd.read_csv("Ha-En_Culture_candidates.tsv", delimiter='\t', header=None)
df.columns=pair
df
```
